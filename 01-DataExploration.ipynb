```python
#Text pre-processing
#1. Normalize the text
#Convert to Lower case
creditcard = creditcard.lower()
```

```python
#Text pre-processing
#3 Remove white spaces
creditcard = creditcard.strip()
```

```python
#Text pre-processing
#4 Remove stop words
#Load the library
from nltk.corpus import stopwords
# Load stop words
stop_words = stopwords.words('english')
set(stopwords.words('english'))
```

```python
#Stop word removal
import nltk
word_tokens = nltk.tokenize.word_tokenize(creditcard) 
  
filtered_sentence1 = [w for w in word_tokens if not w in stop_words] 
  
filtered_sentence1 = [] 
  
for w in word_tokens: 
    if w not in stop_words: 
        filtered_sentence1.append(w) 
  
#print(word_tokens) 
print(filtered_sentence1) 
```

```python
punct = [',', '(', ')', '/', '@', '[', ']', '\\', '{', '}', ';', '-', '_', '&', '.', '"', '``', '·']
```

```python
#Remove Punctuation
filtered_sentence2 = [] 
  
for w in filtered_sentence1: 
    if w not in punct:
        filtered_sentence2.append(w) 
  
#print(word_tokens) 
print(filtered_sentence2) 
```

```python
#Bi-grams - Credit Card
print(list(nltk.bigrams(filtered_sentence2)))
```

```python
fdist = nltk.FreqDist(filtered_sentence2)
print(fdist)
```

```python
import operator
sorted_d = sorted(fdist.items(), key=operator.itemgetter(1),reverse=True)
print(sorted_d[:25])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.title('Frequency Line Chart - Credit Card')
fdist.plot(30, cumulative=False)
```

```python
import pandas as pd
s = pd.DataFrame(list(fdist.items()), columns= ['word', 'count'])
s = s.sort_values(by='count', ascending = False)
s = s.iloc[0:15,:]
s
```

```python
#Frequency Bar Chart - CREDIT CARD
import seaborn as sns
plt.figure(figsize=(20,10))
plt.title('Frequency Bar Chart - Answer Corpus')
sns.barplot(x=s["word"], y=s["count"])
plt.show()
```

```python
comment_words = ' '
for words in filtered_sentence2: 
    comment_words = comment_words + words + ' '
```

```python
#WORD CLOUD - CREDITCARD
from wordcloud import WordCloud, STOPWORDS 
stopwords = set(STOPWORDS) 
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None)
plt.title('Word Cloud - Answer Corpus')
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show() 
```
